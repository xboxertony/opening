```python
import requests
from io import StringIO
import pandas as pd
import numpy as np
import datetime
import time
import matplotlib.pyplot as plt
date = datetime.datetime.now()

def crawler(datestr):
    URL = 'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date='+datestr+'&type=17'
    #headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.2; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
    r = requests.post(URL)
    #r.encoding = 'utf_8_sig'
    df = pd.read_csv(StringIO(r.text.replace('=','')), header=["證券代號" in l for l in r.text.split("\n")].index(True))
    #df = df.iloc[:['備註' in l for l in df['證券代號'].index(True)]
    #df = pd.read_csv(StringIO(r.text),header = 0)
    #df = pd.read_csv(StringIO(r.text))
    
    def k():
        g=0
        for i in range(len(df['證券代號'])):
            g=g+1
            if '備註' in df['證券代號'][i]:
                return g-1
    
    df = df.iloc[:k()]    
    return df

dict_stock = {}
dataspan = 100
date = datetime.datetime.now()
date_index=0

def transfer_date(datestr):
    datestr = str(datestr).split(' ')[0].replace('-','')
    return datestr

while len(dict_stock)<dataspan:
    time.sleep(5)
    now_date = transfer_date(date-datetime.timedelta(days=date_index))
    
    try:
        dict_stock.update({now_date:crawler(now_date)})
    except:
        print('Fails at'+' '+str(transfer_date(now_date)))
    date_index+=1
    
# close = pd.DataFrame({k:d['收盤價'].apply(pd.to_numeric,errors='coerce') for k,d in dict_stock.items()}).transpose()
# close.index = pd.to_datetime(close.index)
# close = close.sort_index()
```
