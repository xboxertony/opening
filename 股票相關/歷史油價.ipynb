```python
import requests as req
import pandas as pd
```

```python
URV = 'https://toolbxs.com/zh-TW/detector/gasoline_price/history_data'
oil_price = req.get(URV)
df = pd.read_html(oil_price.text)
```

```python
oil_grid = df[0]
oil_grid.index = oil_grid[('實施日期','實施日期')]
oil_grid = oil_grid.iloc[::-1,1:]
```

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']  
mpl.rcParams['axes.unicode_minus'] = False

import matplotlib.pyplot as plt
```

```python
oil_grid.index.name = '實施日期'
```

```python
oil_grid
```

```python
plt.figure(figsize=(10,5))

plt.xticks(fontsize=14,rotation = 45)

plt.plot(oil_grid.index,oil_grid.values)
```
