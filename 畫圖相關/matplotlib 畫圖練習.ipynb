```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-3,3,50)
y1=2*x+1
y2=x**2

# plt.figure()
# plt.plot(x,y1)

plt.figure(num=3)                 #長；寬 ,figsize=(8,5)
l1,=plt.plot(x,y2,label='up')     #記得放入handles一定要加逗號
l2,=plt.plot(x,y1,color = 'red',linewidth=1.0,linestyle='--',label='down')
plt.legend(handles=[l1,],labels=['aaa'],loc='best')      #負責顯示標籤

#loc為best他會自動放在好的位置，也可以設定成upper right 等等

plt.xlim((-1,2))
plt.ylim((-2,5))
plt.xlabel('i am x')
plt.ylabel('i am y')

new_ticks = np.linspace(-1,2,5) #負責區間間隔 xticks,yticks
plt.xticks(new_ticks)
plt.yticks([-2,-1,-0.5 ,1.22,2],[r'$really\ bad$',r'$bad\ \alpha$',r'$normal$',r'$good$',r'$really\ good$'])

#gca = 'get current axis'

ax = plt.gca()
ax.spines['right'].set_color('none') #四個邊框 spines
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom') #默認x,y要用哪個軸代替
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0)) #.outward, axes
ax.spines['left'].set_position(('data',0))


x0=1
y0=2*x0+1
plt.scatter(x0,y0,s=50,color='b') #s是size
plt.plot([x0,x0],[0,y0],'k--',lw=2.5) #k是黑色,--虛線,lw線寬度,plot([x0,x1],[y0,y1]) 如果要連接[x0,y0],[x1,y1]

#method
plt.annotate(r'$2x+1=%s$' % y0,xy=(x0,y0),xycoords='data',xytext=(+50,-50),textcoords='offset points'
             ,fontsize=16,arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

# xycoords代表參考值屬性是點還是其他，預設是點:data，xytext代表標籤位置，textcoords代表標籤參考點 +30 代表距離參考點多多少

for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=0.7))
```

```python
n=12
X=np.arange(n)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,facecolor='#9999ff',edgecolor='white')
plt.bar(X,-Y2,facecolor='#ff9999',edgecolor='white')  #edgecolor 邊框顏色
plt.xlim(-1,n)
plt.ylim(-1.25,1.25)

for x,y in zip(X,Y1):
    plt.text(x,y+0.05,'%.2f'%y, ha='center',va='bottom') #ha:horizontal alignment

for x,y in zip(X,-Y2):
    plt.text(x,y-0.1,'%.2f'%y, ha='center',va='top')
    
plt.show()
```

```python
def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

#use plt.contourf to filling contours
#X,Y and value for (X,Y) point 等高線顏色繪製

plt.contourf(X,Y,f(X,Y),10,alpha=0.75,cmap=plt.cm.hot)

#use plt.contourf to add contour lines 等高線繪製
C = plt.contour(X,Y,f(X,Y),10,colors='black',linewidths=1)

plt.clabel(C,inline = True,fontsize=10)

plt.show()
```

```python
%matplotlib notebook
from matplotlib import animation
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()

x=np.arange(0,2*np.pi,0.01)
line,=ax.plot(x,np.sin(x))   #,代表選擇line第一位

def animate(i): #把1~100都丟進去試試看
    line.set_ydata(np.sin(x+i/100))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,func=animate,frames=100,init_func=init,interval=20,blit=True)

#interval 是更新速度，單位是毫秒

plt.show()
```
