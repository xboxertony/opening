```python
from tkinter import *
window = Tk()

l = Label(window,text = 'empty',bg = 'yellow')
l.pack()

def print_selection(v):
    l.config(text = 'you have selected :'+v)

s = Scale(window,label = 'try me'
          ,from_ = 5      #取值範圍
          ,to = 10
          ,orient = HORIZONTAL   #橫向還是豎向
          ,length=200
          ,showvalue = 1
          ,tickinterval=1
          ,resolution = 0.01   #小數點後面幾位
          ,command = print_selection)
s.pack()

window.mainloop()
```

```python
from tkinter import *

window = Tk()
counter=0

l = Label(window,text='')
l.pack()

def do_job():
    global counter
    counter+=1
    l.config(text = 'do'+str(counter))
    

menubar = Menu(window)
filemanu = Menu(menubar,tearoff=0)
menubar.add_cascade(label = 'File',menu = filemanu)
filemanu.add_command(label='New',command = do_job)
filemanu.add_command(label='Open',command = do_job)
filemanu.add_command(label='What?',command = do_job)

submanu = Menu(filemanu)
filemanu.add_cascade(label='Import',menu = submanu)
submanu.add_command(label = 'KKKKK',command = do_job)

filemanu.add_separator()

filemanu.add_command(label='Exit',command = window.quit)


window.config(menu = menubar)

window.mainloop()
```

```python
from tkinter import *
import pickle
from tkinter import messagebox

window = Tk()
window.title('Welcome to Python!!')

#welcome window
canvas = Canvas(window,height=300,width=300)
image_file = PhotoImage(file = r'C:\Users\user\Desktop\ipython\welcome.gif')
image = canvas.create_image(0,0,anchor = 'nw',image = image_file)
canvas.pack(side = 'top')

#Lable
name = Label(window,text = 'User name:').place(x=40,y=170)
password =Label(window,text = 'Password:').place(x=40,y=220)

#Entry
user_name = StringVar()
user_name.set('kkbox.gmail.com')
user_word = StringVar()

entry_name = Entry(window,textvariable = user_name)
entry_name.place(x=110,y=170)
entry_pass = Entry(window,textvariable = user_word,show = '*')
entry_pass.place(x=110,y=220)


#button

def Login_1():
    try:
        with open('user_info.pickle','rb') as user_file:
            user_info = pickle.load(user_file)
    except:
        with open('user_info.pickle','wb') as user_file:
            user_info = {'kkbox.gmail.com':'5566'}
            pickle.dump(user_info,user_file)
    
    if entry_name.get() in user_info.keys():
        if entry_pass.get() == user_info[entry_name.get()]:
            messagebox.showinfo(message = 'Welcome!!!')
        else:
            messagebox.showerror(message = 'Wrong Password!!!')
    else:
        sign_yes_no = messagebox.askyesno('do you want to sigh up?')
        if sign_yes_no:
            Sign_1()
    
def Sign_1():
    
    def sign_up():
        user = entry_name_1.get()
        password = entry_pass_1.get()
        with open('user_info.pickle','rb') as user_file:
            user_info = pickle.load(user_file)
        with open('user_info.pickle','wb') as user_file:
            user_info[user] = password
            pickle.dump(user_info,user_file)
        messagebox.showinfo('You Sign up!!')
        window_sign_up.destroy()
            
    user = StringVar()
    password = StringVar()
    window_sign_up = Toplevel(window)
    window_sign_up.geometry('300x200')
    window_sign_up.title('Sigh up window')
    
    name = Label(window_sign_up,text = 'User name:').place(x=40,y=70)
    password =Label(window_sign_up,text = 'Password:').place(x=40,y=110)
    
    entry_name_1 = Entry(window_sign_up,textvariable = user)
    entry_name_1.place(x=140,y=70)
    entry_pass_1 = Entry(window_sign_up,textvariable = password,show = '*')
    entry_pass_1.place(x=140,y=110)
    
    btn_confirm_sign_up = Button(window_sign_up,text = 'Confirm the account',command = sign_up)
    btn_confirm_sign_up.pack(side = 'bottom')

Button(window,text = 'Login',command = Login_1).place(x=70,y=250)
Button(window,text = 'Sign up',command = Sign_1).place(x=170,y=250)

window.mainloop()
```
