```python
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(i-1)*flip)
```

```python
def f(n,r):
    x = 20
    if n==1:
        x
    else:
        x = f(n-1,r)-r*(2*f(n-1,r)-10)
    return x
```

```python
f(10,0.1)
```

```python
f(20,0.1)
```
