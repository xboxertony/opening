```python
import re,collections
```

```python
def word(text):return re.findall('[a-z]+',text.lower())

def train(features):
    model = collections.defaultdict(lambda:1)
    for f in features:
        model[f]+=1
    return model
```

```python
NWORDS = train(word(open('big5.txt').read()))
```

```python
NWORDS
```

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
```

```python
def edits1(word):
    n = len(word)
    return set([word[0:i]+word[i+1:] for i in range(n)]+
              [word[0:i]+word[i+1]+word[i]+word[i+2:] for i in range(n-1)]+
              [word[0:i]+c+word[i+1:] for i in range(n) for c in alphabet]+
              [word[0:i]+c+word[i:] for i in range(n+1) for c in alphabet])
```

```python
def know_word(word):
    return set(w for w in word if w in NWORDS)
```

```python
def known_edits2(word):
    return set(e2 for e1 in edits1(word) for e2 in edits1(e1) if e2 in NWORDS)
```

```python
def correct(word):
    candidates = know_word([word]) or know_word(edits1(word)) or known_edits2(word) or [word]
    return max(candidates,key = lambda w: NWORDS[w])
```

```python
correct('reasomable')
```
