```python
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
import pandas as pd
```

```python
iris_1 = pd.DataFrame(iris['data'],columns=iris.feature_names)
```

```python
iris_target = pd.DataFrame(iris['target'],columns = ['target'])
```

```python
iris_1 = pd.concat([iris_1,iris_target],axis = 1)
iris_1
```

```python
plt.scatter(iris_1.iloc[:,0],iris_1.iloc[:,2],c=iris_1['target'])
```
