```python
tuples = [(1,'blue','t'),(2,'blue','y'),(2,'red','t'),(3,'red','t'),(3,'blue','r'),(3,'black','t'),(4,'black','r'),(5,'red','r')]
```

```python
columns = pd.MultiIndex.from_tuples(tuples,names=['name','color','type'])

df3 = pd.DataFrame(np.random.randn(5,len(tuples)),columns = columns)
```

```python
df3.sort_values(by = (3,'red','t'),ascending=False)
```

```python
df3.drop(columns=[(1,'blue','t')])
```
